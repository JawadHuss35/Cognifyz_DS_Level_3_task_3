```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming you have the data in a CSV file named 'Dataset .csv'
df = pd.read_csv('Dataset .csv')
df.head(3)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=20, kde=True)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Aggregate rating', data=df)
plt.title('Count of Restaurants by Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
plt.figure(figsize=(30, 6))
sns.barplot(x='Cuisines', y='Aggregate rating', data=df, ci=None)  
plt.title('Average Rating by Cuisine')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Votes', y='Aggregate rating', data=df)
plt.title('Relationship between Votes and Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.show()
```
